```python
#frequency function takes a list as an argument.It computes the frequency of occurence of an object in the list.
#Returns a dictionary with key as the object and values as the frequency of occurence of the object. 

def frequency(list):
    d={}
    for i in list:
        d[i]=d.get(i,0)+1
    return d

#this function calculates the pmf of the random variable.  
def pmf(frequency_dictionary,size):
    pmf={}
    for i in frequency_dictionary:
        pmf[i]=frequency_dictionary[i]/size
    return pmf
```

```python
#this piece of code calculates and plots pmf for the attribute math score.

math=dataset.iloc[:,5].values
math_size=len(math)
math_freq=frequency(math)
#print(math_freq)
math_pmf=pmf(math_freq,math_size)
#print(math_pmf)
p.rcParams['figure.figsize']=40,10
p.bar(math_pmf.keys(),math_pmf.values(),color='c')
p.xticks(list(math_pmf.keys()))
p.show()
p.scatter(math_pmf.keys(),math_pmf.values())
p.show()
```
