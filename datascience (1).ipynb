```python
#read data using pandas
import pandas as pd
dataset=pd.read_csv(r"C:\Users\PreethuPallavi\Documents\dsdata.csv");
```

```python
print(dataset.info())
```

```python
#taking the features as list objects
population=dataset.iloc[:,1].values
ratio=dataset.iloc[:,2].values
state_wise_female_population=[]
#from the gender ratio calculate the actua female population of each state and store in list state_wise_female_population
for i in range(0,37):
   state_wise_female_population.append(round((population[i]*ratio[i])/(ratio[i]+1000)))
print(state_wise_female_population)
```

```python
#finding the total female population of india according to 2011 census
female_population=0
for i in state_wise_female_population:
    female_population+=i;
print(round(female_population))
```

```python
#calculating the probablity mass function 
pmf=[]
for i in range(0,37):                                     
  val=(state_wise_female_population[i]/female_population)
  pmf.append(val)
print(pmf)
```

```python
#verifying the the sum of pmfs is not > 1
sum=0
for i in pmf:
    sum+=i;
print("%d"%sum)
```

```python
#pmf for all states
states=dataset.iloc[:,0].values
import matplotlib.pyplot as p
p.rcParams['figure.figsize']=10,6
p.bar(states,pmf)
p.show()
p.scatter(states,pmf)
p.show()
p.plot(states,pmf)
p.show()
```

```python
#visualizing pmf foe a subset of states
set1=dataset.iloc[30:37,0]
pmf1=pmf[30:37]
import matplotlib.pyplot as p
p.rcParams['figure.figsize']=20,10
p.rc('xtick', labelsize=10)
p.bar(set1,pmf1)
p.show()
p.scatter(set1,pmf1)
p.show()
```
