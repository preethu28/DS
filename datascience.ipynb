```python
import pandas as pd
dataset=pd.read_csv(r"C:\Users\PreethuPallavi\Documents\dsdata.csv");
```

```python
print(dataset.info())
```

```python
population=dataset.iloc[:,1].values
ratio=dataset.iloc[:,2].values
state_wise_female_population=[]
for i in range(0,37):
   state_wise_female_population.append((population[i]*ratio[i])/(ratio[i]+1000))
print(state_wise_female_population)
```

```python
female_population=0
for i in state_wise_female_population:
    female_population+=i;
print(female_population)
```

```python
pmf=[]
for i in range(0,37):                                     
  val=state_wise_female_population[i]/female_population
  pmf.append(val)
print(pmf)
```

```python
sum=0
for i in pmf:
    sum+=i;
print(sum)
```

```python
states=dataset.iloc[:,0].values
```

```python
import matplotlib.pyplot as p
p.bar(states,pmf)
p.figure(figsize=(200,10))
p.show()
p.scatter(states,pmf)
p.show()
p.plot(states,pmf)
p.show()
```
