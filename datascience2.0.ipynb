```python
import pandas as pd
import matplotlib.pyplot as p
dataset=pd.read_csv(r"C:\Users\PreethuPallavi\Desktop\DS DATA\StudentsPerformance.csv")
```

```python
print(dataset.info())
```

```python

def frequency(list):
    d={}
    for i in list:
        d[i]=d.get(i,0)+1
    return d

def pmf(frequency_dictionary,size):
    pmf={}
    for i in frequency_dictionary:
        pmf[i]=frequency_dictionary[i]/size
    return pmf
```

```python
math=dataset.iloc[:,5].values
math_size=len(math)
math_freq=frequency(math)
#print(math_freq)
math_pmf=pmf(math_freq,math_size)
#print(math_pmf)
"""sum=0
for i in math_pmf:
    sum+=math_pmf[i]
print(round(sum))"""

p.bar(math_pmf.keys(),math_pmf.values(),color='c')
p.xticks(list(math_pmf.keys()))
p.rcParams['figure.figsize']=40,10
p.show()

p.scatter(math_pmf.keys(),math_pmf.values())
p.show()
```
